```python
import pandas as pd
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
```

```python
tulips = pd.read_csv('./data/tulips.csv', sep=';')
```

```python
water = tulips.water - tulips.water.mean()
shade = tulips.shade - tulips.shade.mean()
blooms = tulips.blooms - tulips.blooms.max()
```

```python
with pm.Model() as tulips3:
    beta_water = pm.Uniform('beta_water',0,10)
    beta_shade = pm.Uniform('beta_shade',-10, 0)
    beta_interaction = pm.Normal('beta_interaction',0,0.25)
    """
    The motivation to choose this prior is that
    we believe that a lot of shade makes water have an oposite effect, but proportional, so we set the same
    sd as beta_shared
    """
    alpha = pm.Normal('alpha',0.5,0.25)
    sigma = pm.Exponential('sigma',1)
    mu = pm.Deterministic('mu', alpha+beta_shade * shade + beta_water * water + beta_interaction * water * shade)
    y = pm.Normal('y',mu, sigma, observed=blooms)
```

```python
rugged = pd.read_csv('./data/rugged.csv', sep=';')
rugged = rugged.dropna(axis=0, subset=["rgdppc_2000"])
rugged["log_gdp"] = np.log(rugged["rgdppc_2000"])
rugged["s_log_gdp"] = rugged["log_gdp"] / np.mean(rugged["log_gdp"])
rugged["s_rugged"] = rugged["rugged"] / np.max(rugged["rugged"])
```

8H3 robust regression

```python
with pm.Model() as m3:
    is_african_country = pm.MutableData("is_african_country", rugged.cont_africa)
    ruggedness = pm.MutableData("ruggedness",rugged.s_rugged)
    alphas = [pm.Normal('alpha_africa', 1, 0.1), pm.Normal('alpha_nonafrica', 1, 0.2)]
    betas = [pm.Normal('beta_africa', 0, 0.3), pm.Normal('beta_nonafrica', 0, 0.3)]
    alpha = alphas[0] * is_african_country + alphas[1] * (1-is_african_country)
    beta = betas[0] * is_african_country + betas[1] * (1-is_african_country)
    mu = pm.Deterministic('mu', alpha + beta * (ruggedness-0.215))
    sigma = pm.Exponential('sigma',1)
    y = pm.Normal('y', mu=mu, sigma=sigma, observed=rugged.s_log_gdp)
```

```python
with m3:
    m3_trace = pm.sample()
```

```python
def posterior_prediction_plots(posterior_predictive_african, posterior_predictive_non_african):
    fig, axs = plt.subplots(1, 2, figsize=(15, 7))
    fig.suptitle("Posterior prediction plots", fontsize=15)
    africa = rugged[rugged.cont_africa==1]
    non_africa = rugged[rugged.cont_africa==0]

    # scatter sample values
    axs[0].scatter(africa.s_rugged, africa.s_log_gdp)
    axs[1].scatter(non_africa.s_rugged, non_africa.s_log_gdp)

    #simulate predictions over X axis
    az.plot_hdi(x,posterior_predictive_african.posterior_predictive['y'], ax=axs[0])
    az.plot_hdi(x,posterior_predictive_non_african.posterior_predictive['y'], ax=axs[1])

    #simulate mu
    az.plot_hdi(x,posterior_predictive_african.posterior_predictive['mu'], ax=axs[0])
    az.plot_hdi(x,posterior_predictive_non_african.posterior_predictive['mu'], ax=axs[1])
    
    #the median mu
    axs[0].plot(x, np.vstack(posterior_predictive_african.posterior_predictive["mu"].to_numpy()).mean(0))
    axs[1].plot(x, np.vstack(posterior_predictive_non_african.posterior_predictive["mu"].to_numpy()).mean(0))

    axs[0].set_xlabel('Standarized (0-1) Ruggedness')
    axs[0].set_ylabel('Standarized Log GDP')
    
posterior_prediction_plots(m3_posterior_pred_african, m3_posterior_pred_non_african)
```

```python
def robust_vs_non_robust(posterior_predictive_african,
                         posterior_predictive_non_african,
                        robust_posterior_predictive_african,
                        robust_posterior_predictive_non_african):
    fig, axs = plt.subplots(1, 2, figsize=(15, 7))
    fig.suptitle("Posterior prediction plots", fontsize=15)
    for axis, countries, post, robust_post in [(0, rugged[rugged.cont_africa==1], posterior_predictive_african, robust_posterior_predictive_african),
                       (1, rugged[rugged.cont_africa==0], posterior_predictive_non_african, robust_posterior_predictive_non_african)]:

        # scatter sample values
        axs[axis].scatter(countries.s_rugged, countries.s_log_gdp)
        az.plot_hdi(x,post.posterior_predictive['mu'], ax=axs[axis])
        az.plot_hdi(x,robust_post.posterior_predictive['mu'], ax=axs[axis])
        axs[axis].plot(x, np.vstack(post.posterior_predictive["mu"].to_numpy()).mean(0))
        axs[axis].plot(x, np.vstack(robust_post.posterior_predictive["mu"].to_numpy()).mean(0))    
        axs[axis].set_xlabel('Standarized (0-1) Ruggedness')
        axs[axis].set_ylabel('Standarized Log GDP')
```

```python
# Want to understand the effects of mean.growing.season and sd.growing.season
Want to understand if there's any interaction between mean.growing.season and sd.growing season. 
Possible the higher the mean.growing.season, the lower its sd.?
A way to think about it sd. is uncertainty with respect to how long the growing season takes.

```

```python
with pm.Model() as nettle_m1:
    mutable_area = pm.MutableData('mutable_area',value=nettle_areas)
    mutable_mgs = pm.MutableData('mutable_mgs',value=nettle['mean.growing.season'])
    beta_area = pm.Normal('beta_area', 0,1)
    beta_mean_growing_season = pm.Normal('beta_mean_growing_season', 0,1)
    alpha = pm.Normal('alpha', 0,1)
    sigma = pm.Exponential('sigma',1)
    mu = pm.Deterministic('mu', alpha + beta_area * mutable_area + beta_mean_growing_season * mutable_mgs)
    y = pm.Normal('y', mu=mu, sigma=sigma, observed=log_lang_per_capita)
```

```python
with pm.Model() as nettle_m2:
    mutable_area = pm.MutableData('mutable_area',value=nettle_areas)
    mutable_sd = pm.MutableData('mutable_sgs',value=nettle['sd.growing.season'])
    
    beta_area = pm.Normal('beta_area', 0,1)
    beta_sd_growing_season = pm.Normal('beta_sd_growing_season', 0,1)
    
    alpha = pm.Normal('alpha', 0,1)
    sigma = pm.Exponential('sigma',1)
    mu = pm.Deterministic('mu', alpha + beta_area * mutable_area + beta_sd_growing_season * mutable_sd)
    y = pm.Normal('y', mu=mu, sigma=sigma, observed=log_lang_per_capita)
```

```python
with pm.Model() as nettle_m3:
    mutable_area = pm.MutableData('mutable_area',value=nettle_areas)
    mutable_mgs = pm.MutableData('mutable_mgs',value=nettle['mean.growing.season'])
    mutable_sd = pm.MutableData('mutable_sgs',value=nettle['sd.growing.season'])
    beta_area = pm.Normal('beta_area', 0,1)
    beta_mean_growing_season = pm.Normal('beta_mean_growing_season', 0,1)
    beta_sd_growing_season = pm.Normal('beta_sd_growing_season', 0,1)
    beta_interaction = pm.Normal('beta_interaction', 0,1)
    alpha = pm.Normal('alpha', 0,1)
    sigma = pm.Exponential('sigma',1)

    mu = pm.Deterministic('mu', alpha + beta_area * mutable_area + beta_mean_growing_season * mutable_mgs+ beta_sd_growing_season*mutable_sd + beta_interaction * mutable_mgs * mutable_sd)
    y = pm.Normal('y', mu=mu, sigma=sigma, observed=log_lang_per_capita)
```

```python
with pm.Model() as nettle_m4:
    mutable_area = pm.MutableData('mutable_area',value=nettle_areas)
    mutable_mgs = pm.MutableData('mutable_mgs',value=nettle['mean.growing.season'])
    mutable_sd = pm.MutableData('mutable_sgs',value=nettle['sd.growing.season'])
    beta_area = pm.Normal('beta_area', 0,1)
    beta_mean_growing_season = pm.Normal('beta_mean_growing_season', 0,1)
    beta_sd_growing_season = pm.Normal('beta_sd_growing_season', 0,1)
    alpha = pm.Normal('alpha', 0,1)
    sigma = pm.Exponential('sigma',1)

    mu = pm.Deterministic('mu', alpha + beta_area * mutable_area + beta_mean_growing_season * mutable_mgs+ beta_sd_growing_season*mutable_sd)
    y = pm.Normal('y', mu=mu, sigma=sigma, observed=log_lang_per_capita)
```

```python
Based on our plots
1. Looking at MGS alone, it seems to be decremental to LPC
2. Looking at MGS stratified by area (country size), we can consistenly perceive the oppositve behavior, it affects 
possitevely LPC.
3. Looking at area alone, it seems to be decremental to LPC

For now, what I think is going on here is that the influence of MGS is positive. When looking at MGS without stratifing
we are getting confounded by log(area), which forms a backdoor path 

Dag:
    log(area) -> MGS -> LPC
    log(area) -> LPC.

I guess also that there could be some interaction between area and pop with LPC. If pop is low and area is big, then
there could be an increase in dispersion thus in languages? if pop increases, we have more connection, if area increases 
we have less connection. This is just a hypothesis, no domain expert here.
```

```python
judge_to_index = {judge:i for i, judge in enumerate(wines.judge.unique())}
wine_to_index = {wine:i for i, wine in enumerate(wines.wine.unique())}
```

```python
numeric_judges = wines.judge.apply(lambda x :judge_to_index[x])
numeric_wines = wines.wine.apply(lambda x :wine_to_index[x])
```
